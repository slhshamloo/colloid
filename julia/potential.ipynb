```julia
using Plots, Colors, ColorSchemes, Statistics, CSV, DataFrames, JLD2, CUDA
Plots.default(fontfamily="Computer Modern")
include("src/HPMC.jl")
```

```julia
ds = collect(0.0:0.25:4.0)
katics = Vector{Float32}(undef, length(ds))
errors = Vector{Float32}(undef, length(ds))
for (i, d) in enumerate(ds)
    df = DataFrame(CSV.File("data/katic4gon4.0r$(d)d200.0p.csv"))
    katics[i] = mean(df.katic)
    errors[i] = std(df.katic) / (length(df.katic))
end
potentials = maximum(katics) .- katics
p = plot(ds, potentials, yerr=errors, xlabel="\$d\$", ylabel="Potential",
    lw=2, color=:blue, msc=:blue, label="\$P=200.0\$")

ds = collect(0.0:0.25:4.0)
katics = Vector{Float32}(undef, length(ds))
errors = Vector{Float32}(undef, length(ds))
for (i, d) in enumerate(ds)
    df = DataFrame(CSV.File("data/katic4gon4.0r$(d)d100.0p.csv"))
    katics[i] = mean(df.katic)
    errors[i] = std(df.katic) / (length(df.katic))
end
potentials = maximum(katics) .- katics
plot!(p, ds, potentials, yerr=errors, xlabel="\$d\$", ylabel="Potential",
    lw=2, color=:red, msc=:red, label="\$P=100.0\$")
```
